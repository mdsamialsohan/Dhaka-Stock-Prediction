```python
import matplotlib.pyplot as plt

# Data
features = ["Basic Features", "EMA", "EMA, SMA-7, MACD", "Ensemble Method"]
durations = ["1 Day", "1 Week", "15 Days"]
data = {
    "Basic Features": [4.835, 8.914, 13.91025],
    "EMA": [4.28425, 9.22125, 13.14275],
    "EMA, SMA-7, MACD": [4.265, 8.7155, 14.03825],
    "Ensemble Method": [3.766, 7.455, 11.699]
}

# Create a bar chart
num_features = len(data)
width = 0.2  # Width of each bar, adjusted to fit multiple bars side-by-side
index = range(len(durations))  # Index for bars

plt.figure(figsize=(10, 6))  # Set the figure size

# Create bars for each feature, shifting the position for each feature
for i, (feature, values) in enumerate(data.items()):
    offset = [x + width*i for x in index]  # Shift the position for each feature
    plt.bar(offset, values, width, label=feature)

# Set labels and title
plt.xlabel("Duration")
plt.ylabel("RMSE")
plt.title("Comparison of Ensemble Method with other model")

# Set x-axis ticks and labels
# Adjust the ticks to be in the center of the groups
plt.xticks([x + width*(num_features/2 - 0.5) for x in index], durations, rotation=45, ha="right")

# Add legend
plt.legend()

# Show the chart
plt.tight_layout()
plt.show()
```
